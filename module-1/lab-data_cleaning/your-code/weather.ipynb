## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import numpy as np
import pandas as pd

df = pd.read_csv("../weather-raw.csv")
```

```python
dff = df.fillna(0)
dff
```

```python
grouped = dff.groupby("month").sum()
grouped.head()
```

```python

cut_df = df[df.columns[4:35]]
df_sum = cut_df.sum()
df_count = cut_df.count(axis=0)
df_mean = df_sum/df_count
print("media diaria durante todos los meses")
print(df_mean)
```

```python
month = grouped.T[1:32]
month_sum = month.sum()
month = month.replace(0.0, np.nan)
month_count = month.notnull().sum()
month_mean = month_sum/month_count
#print("media mensual: ",month_mean)


month = dff.T
month =  month[4:35]
display(month)
month_sum = month.sum()
month = month.replace(0.0, np.nan)
month_count = month.notnull().sum()

final_count = []
final_sum = []
for i in range(0,22,2):
    print(i)
    final_count.append(month_count[i]+month_count[i+1])
    final_sum.append(month_sum[i]+month_sum[i+1])

result=[]
for i in range(len(final_sum)):
    result.append(final_sum[i]/final_count[i])

print("media mensual:")
for r in range(len(result)):
    print("\nMonth",r+1)
    print(result[r])

```
